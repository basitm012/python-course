Q1. Create a dictionary called student with keys: name, age, course, city. Print each value.

```python
students = {
    'Name': 'Yousuf',
    'Age': 15,
    'Course': 'Computer',
    'City': 'Karachi'
 }
```

```python
print(students)

print(students['Name'])
print(students['Age'])
print(students['Course'])
print(students['City'])
```
