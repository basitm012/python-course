## **Dictionaries**

Dictonaries are key-value points

```python
bio_data_list = ["Abdul", "5.5", 17, "Clifton, Karachi."]

bio_data_dic =  {
    'Name' : 'Abdul Basit',
    'Height' : '5.5',
    'Age' : 17,
    'Address' : 'Clifton, Karachi.'
}
```

```python
bio_data_dic
```

```python
bio_data_list [0] #extract value is list
bio_data_dic ['name'] #extract value is dictionary
```

```python
bio_data_list.append('0300-1234567') # add value in list
bio_data_dic ['Phone'] = '0300 1234567' # add value in list
```
